<h3>Step 1: Importing the required libraries </h3>

<p>This step involves just importing the required libraries which are numpy, pandas and csv. The reason behind importing csv is that the dataset which we are about to with is of csv type, so this is the reason for importing csv.</p>

```python
import pandas as pd
import numpy as np
import csv
```

<h3>Step 2: Getting the dataset from a different source and displaying the dataset</h3>

<p>This step involves getting the dataset from a different source, and the link for the dataset is provided below, the head(5) is used to display the top 5 rows of the dataset.</p>

[Dataset download](https://www.kaggle.com/ronitf/heart-disease-uci)

```python
df = pd.read_csv("C:/Users/Tanu/Desktop/Datasets/heart.csv")
df.head(5)
```

<h3>Step 3: Removing the unused or irrelevant columns</h3>

<p>This step involves removing irrelevant coloumns such as cp, fbs, thalach and many more, and the code is pretty much self-explanatory.</p>

```python
to_drop = ['cp',
          'fbs',
          'restecg',
          'thalach',
          'exang',
          'oldpeak',
          'slope',
          'thal',
           'target', 
          'ca']

df.drop(to_drop, inplace = True, axis = 1)
df.head(5)
```

<h3>Step 4: Renaming the coloumn names as per our convenience</h3>

<p>This step involves renaming the columns names because many column names are kinda confusing and hard to understanding.</p>

```python
new_name = {'age': 'Age',
           'sex': 'Sex',
           'trestbps': 'Bps',
           'chol': 'Cholesterol'
            }

df.rename(columns =new_name, inplace = True)
df.head()
```

<h3>Step 5: Replacing the value of the rows if necessary</h3>

<p>This step involves replacing the incomplete values or making the values understandable, such as the Sex field consists of the data 1 and 0 being 1 as Male and 0 as Female, but it often seems ambiguous for the third person, so changing the value to a understandable one is a good idea.</p>

```python
replace_values = {0: 'F', 1: 'M'}

df = df.replace({"Sex": replace_values})                                                                                             

df.head()
```
